```python
import pandas as pd
import folium as g
df = pd.read_csv('GPS_D.csv', encoding='cp949')
lines = df.iloc[:,2:4].values.tolist()
#time = df['시간'].tolist()
```

```python
lines
```

```python
# 시작위치
import numpy as np

start,end = lines[0],lines[-1]
gps_m = g.Map(location = start, zoom_start = 12)

g.PolyLine(locations = lines).add_to(gps_m)
start_point = g.Marker(start,
                  popup = 'start_point',
                  icon = g.Icon(color='blue'))
end_point = g.Marker(end,
                  popup = 'end_point',
                  icon = g.Icon(color='black'))
start_point.add_to(gps_m)
end_point.add_to(gps_m)
```
