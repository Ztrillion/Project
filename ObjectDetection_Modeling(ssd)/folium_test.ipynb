```python
import os

path = './GPS_FILE/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] 
```

```python
import pandas as pd
import folium as g
df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i)
    gps_df = pd.concat([df,data], ignore_index=True)
    
gps_df = df.drop(columns='1')
```
